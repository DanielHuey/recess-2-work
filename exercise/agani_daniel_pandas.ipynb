```python
def d():
    return f"\n{'-':->30}"
```

```python
s1 = pd.Series([4, 65, 436, 3, 9])
s2 = pd.Series([7, 0, 3, 897, 9])
print(s1.corr(s2),d())
print(s1==s2,d())
```

```python
s1 = pd.Series([2,4,6,8,14])
s2 = pd.Series([1,3,5,7,9])

print("Add")
print(s1+s2,d())
print("Sub")
print(s1-s2,d())
print("Mul")
print(s1*s2,d())
print("Div")
print(s1/s2,d())
```

```python
def strip_list(list):
    for i in range(len(list)):
        list[i] = list[i].strip()
    return list
```

```python
af['AwayTeam'] = strip_list(af['AwayTeam'])
```
